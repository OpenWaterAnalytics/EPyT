### Loads 2 different networks.
 
This example contains:
* Load networks.
* Print elevations for the two networks.
* Plot networks.
* Unload libraries.

```python
from epyt import epanet

# Load networks
d1 = epanet('Net1.inp')
d2 = epanet('Net2.inp')
```

```python
# Print elevations for the two networks.
print('Net1 - Elevations:')
print('------------------')
print(d1.getNodeElevations())
print('Net2 - Elevations:')
print('------------------')
print(d2.getNodeElevations())

# Set new elavations and print them.
d1.setNodeElevations(1, 750.22)
d2.setNodeElevations(2, 200.33)

print('Net1 - New Elevations:')
print('------------------')
print(d1.getNodeElevations())
print('Net2 - New Elevations:')
print('------------------')
print(d2.getNodeElevations())
```

```python
# Print computed values for the two networks.
print('Net1 - Computed Time Series Values:')
print('-----------------------------------')
d1_comp_vals = d1.getComputedTimeSeries()
print(f'Flow: {d1_comp_vals.Flow}')
print(f'Velocity: {d1_comp_vals.Velocity}')
print('\nNet2 - Computed Time Series Values:')
print('-----------------------------------')
d2_comp_vals = d2.getComputedTimeSeries()
print(f'Flow: {d2_comp_vals.Flow}')
print(f'Velocity: {d2_comp_vals.Velocity}')
```

```python
# Plot networks.
d1.plot()
d2.plot()
```

```python
# Unload libraries.
d1.unload()
d2.unload()
```
