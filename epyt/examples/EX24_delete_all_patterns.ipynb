### Delete all patterns from network file

This example contains:
* Load network.
* Delete all patterns.
* Save new file without patterns.
* Unload library.

```python
import os 
os.chdir("../../")
from epyt import epanet

# Load network
d = epanet('Net2.inp')
```

```python
# Delete all patterns
print(f"Pattern indices: {d.getPatternIndex()}")

while d.getPatternIndex():
    indexPat = d.getPatternIndex()
    d.deletePattern(indexPat[0])

# Save new file without patterns
d.saveInputFile('No_patterns.inp')
```

```python
# Print the new file 'No_patterns.inp'
no_pat_inp = open('No_patterns.inp', "r")
print(no_pat_inp.read())
```

```python
# Unload library
d.unload()
```
